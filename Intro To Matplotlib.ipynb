```python
plt.plot([1,2,3] ,[4,5,6])
plt.show()
```

```python
#Data Example
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
plt.show()
```

```python
plt.plot(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years, kohli , label='Virat Kohli')
plt.plot(years, sehwag, label="Virender Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli, 'ro--', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'c*:' , label="Sehwag")  
plt.legend()
plt.show()
#if linestyle and marker not mentioned separately then we cannot change its size also so better to write in readdable format(my tip)
```

```python
#format string in readdable format (another format with same result)
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle=':', marker= "*" , ms=10 , label="Sehwag")
plt.plot(years, runs, color='blue', linewidth=3, label="Tendulkar")
plt.legend()
plt.grid(True) #for grid
```
