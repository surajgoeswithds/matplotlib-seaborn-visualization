```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
import numpy as np
x = np.arange(len(years)) # as year has 10 values only
```

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width = 0.25

plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")

# Each year has one position on the x-axis (0, 1, 2... from np.arange). If you plot all three players at the same position, the bars stack on top of each other — you can't see them separately.
# So you shift each player's bar left or right by width amount:

# Sachin goes at x - width (shifted left)
# Sehwag stays at x (center)
# Kohli goes at x + width (shifted right)

# That's why you get three bars sitting side by side for each year instead of overlapping.
# And np.arange gives you clean integer positions (0, 1, 2...) so the math works. If you used actual years (1990, 1992...) the spacing would be inconsistent.

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,... {We use plt.xticks() to set the correct category labels like years or names.}
plt.legend()
# plt.tight_layout()
# plt.show()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color="purple") # we can use hex value of colours too

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 50, str(runs[i]), ha='center') #Tip: Use ha='center' to center the text and add a small offset (+50 here) to avoid overlap with the bar.

plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
plt.show()
```

```python
# Summary
# Feature	Use
# plt.bar()	Vertical bars for categorical comparison
# plt.barh()	Horizontal bars (great for long labels)
# width=	Control thickness/spacing of bars
# np.arange()	Helps to align multiple bars side by side
# plt.xticks()	Replaces index numbers with real labels
# plt.tight_layout()	Prevents labels from overlapping
```
